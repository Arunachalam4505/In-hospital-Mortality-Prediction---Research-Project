```python
#bootstrapping confidence intervals

a= test[test['mortality30days']==1]
b= test[test['mortality30days']==0]


a=a.to_numpy()
b=b.to_numpy()
```

```python

    
n_size=len(a)
n_size1=len(b)
    
n_iterations = 2000
stats=[]

for i in range(n_iterations):       
    test1 = sklearn.utils.resample(a, n_samples=n_size,replace=True)
    test2 = sklearn.utils.resample(b, n_samples=n_size1,replace=True)

    test= np.concatenate((test1,test2))
    ytest= test[:,0]
    y=test[:,1:40]
    predictions = model.predict(y)

    score = roc_auc_score(ytest, predictions)
    print('auc_score',score)
    stats.append(score)
        
pyplot.hist(stats)
pyplot.show()
alpha = 0.95
p = ((1.0-alpha)/2.0) * 100
    
import numpy
lower = max(0.0, numpy.percentile(stats, p))
p = (alpha+((1.0-alpha)/2.0)) * 100
upper = min(1.0, numpy.percentile(stats, p))
print('%.1f confidence interval %.1f%% and %.1f%%' % (alpha*100, lower*100, upper*100)) 
print('Average Bootstrapped Scores=%.3f',sum(stats)/n_iterations)
```

```python
start = time.time()
space_lr ={
        'max_depth': hp.randint('max_depth', 5, 70),
        'max_features': hp.choice('max_features', ['sqrt', 'log2']),
        'criterion': hp.choice('criterion', ["gini", "entropy"]),
             'min_samples_leaf': hp.choice('min_samples_leaf',[1,100,200,300,400,500])}

def dt(params):
    params = {'max_depth': params['max_depth'],
                  'min_samples_leaf': params['min_samples_leaf'],
                  'criterion': params['criterion'],
                 'max_features': params['max_features']}


    dt_bo = sklearn.tree.DecisionTreeClassifier(random_state=42, **params)
    best_score = cross_val_score(dt_bo, X_train, y_train, scoring='roc_auc', cv=5).mean()
    return 1-best_score
    
dt_best_param = fmin(fn=dt,
                    space=space_lr,
                    max_evals=24,
                    rstate=np.random.RandomState(42),
                    algo=tpe.suggest)
print('It takes %s minutes' % ((time.time() - start)/60))
print('Decision Tree Best Parameters',dt_best_param)
    
```

```python
param_values = [x for x in dt_best_param.values()]

if param_values[2] ==0:
    max_features = 'sqrt'
else:
    max_features ='log2'

if param_values[0] == 0:
    criterion = 'entropy'
else:
    criterion= 'gini'
        
max_features=param_values[2]

dt_best =sklearn.tree.DecisionTreeClassifier(
    criterion=criterion,
     max_depth= param_values[1],
    min_samples_leaf= param_values[3],
    max_features=max_features)

                                   
model= dt_best.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
space_lr = {'max_depth': hp.choice('max_depth', range(5, 70,1)),
                'learning_rate':hp.quniform('learning_rate',0.01,0.5,0.01),
            'n_estimators':hp.choice('n_estimators',range(20,200,5)),
            'gamma':hp.quniform('gamma',0,0.5,0.01),
            'min_child_weight':hp.quniform('min_child_weight',1,10,1),
            'subsample':hp.quniform('subsample',0.1,1,0.01),
            'colsample_bytree':hp.quniform('colsample_bytree',0.1,1,0.01)
            
        }


def xgb(params):
    params = {'max_depth': params['max_depth'],
                 'learning_rate': params['learning_rate'],
                  'gamma':params['gamma'],
                  
                'min_child_weight':params['min_child_weight'],
              'subsample':params['subsample'],
                  'n_estimators': params['n_estimators'],
              'colsample_bytree':params['colsample_bytree']
                  
                 }
    
    xgb_bo = XGBClassifier( **params)

    best_score = cross_val_score(xgb_bo, X_train, y_train, scoring='roc_auc', cv=5).mean()
    return 1 - best_score
xgb_best_param = fmin(fn=xgb,
                    space=space_lr,
                    max_evals=24,
                    rstate=np.random.RandomState(42),
                    algo=tpe.suggest)
print(xgb_best_param)
```

```python
param_values = [x for x in xgb_best_param.values()]


xgb_best = XGBClassifier(colsample_bytree= param_values[0],
   gamma= param_values[1],
    learning_rate= param_values[2],
   max_depth= param_values[3],
   min_child_weight=param_values[4],
n_estimators = param_values[5],
subsample=param_values[6]
   )

xgb_best=XGBClassifier()

model= xgb_best.fit(X_train, y_train)
pred = model.predict(X_test)

```

```python
#Random Forest Classifier


start = time.time()
space_lr = {'max_depth': hp.randint('max_depth', 3, 50),
                'max_features': hp.randint('max_features', range(1,150,10)),
                'n_estimators': hp.choice('n_estimators', range(100,500)),
                'criterion': hp.choice('criterion', ['entropy','gini']),
                    }

def rf(params):
    params = {'max_depth': params['max_depth'],
                  'max_features': params['max_features'],
                  'n_estimators': params['n_estimators'],
                  'criterion': params['criterion'],
             }
    rf_bo = sklearn.ensemble.RandomForestClassifier(n_jobs=-1, **params)
    best_score = cross_val_score(rf_bo, X_train, y_train, scoring='roc_auc', cv=5).mean()
    return 1-best_score

rf_best_param = fmin(fn=rf,
                    space=space_lr,
                    max_evals=24,
                    rstate=np.random.RandomState(42),
                    algo=tpe.suggest,
                        )
print('It takes %s minutes' % ((time.time() - start)/60))
print('Random Forest Best Parameters',rf_best_param)
```
